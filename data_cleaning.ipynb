```python
daily_df['day'] = pd.to_datetime(daily_df['day'], format='%Y-%m-%d', errors='ignore')
daily_df.head()
```

```python
print(daily_df['day'].isna().sum())
print(daily_df['energy_sum'].isna().sum())
print(daily_df['LCLid'].isna().sum())
```

```python
daily_df.dropna(inplace=True, subset=['energy_sum'])
```

```python
holidays_df['Bank holidays'] = pd.to_datetime(holidays_df['Bank holidays'], format='%Y-%m-%d', errors='ignore')
holidays_df.dtypes
```

```python
holidays_df['Bank holidays']
```

```python
daily_df['is_holiday'] = 0
for holiday in holidays_df['Bank holidays']:
  daily_df.loc[daily_df['day'] == holiday, 'is_holiday'] = 1
```

```python
daily_df['is_holiday'].value_counts()
```

```python
weather_df['temperatureMaxTime'] = pd.to_datetime(weather_df['temperatureMaxTime'], utc=True)
weather_df.dtypes
```

```python
weather_df['day'] = weather_df['temperatureMaxTime'].apply(lambda x: x.date)
weather_df['day'] = pd.to_datetime(weather_df['day'], format='%Y-%m-%d', errors='ignore')
weather_df.dtypes
```

```python
weather_df_cleaned = weather_df[['temperatureMin', 'temperatureMax', 'humidity', 'cloudCover', 'windSpeed',
                                 'day','dewPoint','pressure','uvIndex']]
weather_df_cleaned
```

```python
daily_df.dtypes
```

```python
daily_df = daily_df.merge(weather_df_cleaned, how='left', on='day')
daily_df.head(10)
```

```python
daily_df['is_weekend'] = daily_df['day'].apply(lambda x: 1 if x.isoweekday()==6 or x.isoweekday()==7 else 0)
daily_df
```

```python
daily_df = daily_df.merge(household_df, how='left', on='LCLid')
daily_df
```
